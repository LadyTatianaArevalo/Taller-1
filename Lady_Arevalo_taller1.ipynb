### **Imagen 2. - Distribución de casos de covid19 y municipios vecinos gobierno de México 2020**

La gráfica 2, muestra el estado de contagios en México para abril de 2020. En la imagen se visualiza la distribución de casos de covid-19 por los municipios en el mapa político de México, donde se observan 3 categorías del estado de la población mexicana (Con contagios, vecinos de contagios y sin contagios ni vecindad).


* **¿qué?**

¿Qué datos ve el usuario?

Los datos que se pueden visualizar en la gráfica, corresponde a datos de tipo discreto, categóricos y espacial, dado que representa la cantidad de contagios identificados en los municipios de México. Por la naturaleza de los datos, es posible asumir que su tratamiento puede asemejarse a una distribución binomial, donde se obtiene la cantidad de casos positivos/negativos de covid-19.

¿Qué tipo de datos se proporcionan?

Los datos que se muestran en la gráfica son datos categóricos, que proporcionan información de los contagiados y también de personas cercanas (vecinos) a estos casos positivos y la población sin contagio y sin cercanía de contagios. Estas categorizaciones ayudan a controlar la propagación del virus al considerar la población de alto riesgo y bajo riesgo.

¿Qué información puedes deducir de los datos, frente a los significados que te deben decir explícitamente?

En la gráfica de la distribución de casos covid-19, se puede identificar las zonas de México con mayor nivel de contagio o que pueden representar por ejemplo una ocupación alta de hospitalización o cuidados intensivos. Por supuesto, también se puede observar áreas donde no se presentan contagios hasta abril del 2020.

* **¿por qué?** 

¿Qué tipo de tarea requiere el usuario?

Con un gráfico de estos un usuario como una entidad de control o gubernamental espera recibir información de las zonas de México y su positividad del virus, para desarrollar políticas públicas que garanticen de alguna forma la desaceleración de propagación del virus. Como también, conocer municipios o ciudades con bajas tasas de contagio, para implementar protocolos que ayuden a mantener bajos contagios.

¿Requiere el usuario realizar transformaciones en los datos?

Además de conocer las cantidades de población contagiada y sin contagio, también se deben analizar información técnica epidemiológica donde no necesariamente se debe hacer una transformación a los datos, pero sí conocer por ejemplo el número RO; el cual indica la probabilidad de contagio de una persona positiva en número de personas posibles a contagiar.


* **Marcas y canales utilizados**


-	Mapa del Estado de México con datos de contagios de Covid-19
-	Canal de posición espacial
-	Canal de color, por cada categoría relacionada con los contagios
